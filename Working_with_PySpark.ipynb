```python
from pyspark.sql import SparkSession
```

```python
# create a spark session
spark = SparkSession.builder.appName('Read CSV').getOrCreate()
```

```python
# Path to your csv file
csv_file_path = '/content/drive/MyDrive/data.csv'
```

```python
csv_file_path
```

```python
# Read the CSV file into a DataFrame. inferSchema tries to determine the datatype of values in the fields.
df = spark.read.csv(csv_file_path, header=True, inferSchema=True)
```

```python
# Adding columns to the Dataframe

df = df.withColumn('Experience after 2 years', df['Experience'] + 2)
```

```python
# Drop enteries that have all columns as null

df.na.drop(how="all")
df.show()
```

```python
df.na.drop(how="any").show()
```

```python
df.show()
```

```python
df = df.na.drop(how="any", thresh=2)
df.show()
```

```python
# Drop rows that has Experience as NULL

df = df.na.drop(how="any", subset=['Experience'])
df.show()
```

```python
df.show()
```

```python
df = df.na.fill(0, ['Age', 'Experience'])
df.show()
```

```python
# fill in the said text in the string columns that are NULL
df = df.na.fill("Missing")
df.show()
```

```python
from pyspark.ml.feature import Imputer
```

```python
df.show()
```

```python
imputer = Imputer(
inputCols = ['Age','Experience'], outputCols = ["{}_imputed".format(c) for c in
['Age','Experience']]).setStrategy("mean")

imputer.fit(df).transform(df).show()
```
